```python
uk=['schools','mortgageRates','address','resoFacts','listingSubType']
df=df.drop(uk,axis=1)

#Property type dimension will already be dropped, because it is a 1 to 1 copy of hometype
#living Area Value is a 1 to 1 coyp to living Area
#dropping before the analysis is better then removing na's that take away data from other rows
rd=['propertyTypeDimension','livingAreaValue']
df=df.drop(rd,axis=1)
#ptd = room type
#lav = living area
```

<h4>Converting and Encoding</h4>


```python
#zip code will prob need to be dropped, and cant be used for ml since its suggesting a hirachical order, wich zip codes dont have (they kinda have a geographical hirachcal order, but i do not think the ml can understand that)
```

```python
def convert_price(df):
    # Apply the conversion for rows where the currency is 'CAD'
    df.loc[df['currency'] == 'CAD', 'price'] *= 0.74
    return df

# Since most of the hoa fee is 0, ill assume, if Na= 0
df['monthlyHoaFee'] = df['monthlyHoaFee'].fillna(0)

# Extract the "days" from 'timeOnZillow' and convert to float
df['timeOnZillow'] = df['timeOnZillow'].str.extract('(\d+)').astype(float)
    
```

```python
# Hometype: Unknow will be counted as Na, sicne there are only 3 we drop them
df = df[df['homeType'] != 'HOME_TYPE_UNKNOWN']

# this makes multiple columns with the variable (Separate for yes/no)
from sklearn.preprocessing import OneHotEncoder
variables = ['homeType']

# use encoder
encoder = OneHotEncoder(sparse_output=False).set_output(transform="pandas")
one_hot_encoded = encoder.fit_transform(df[variables]).astype(int)
df = pd.concat([df,one_hot_encoded],axis=1).drop(columns=variables)

# homeType_TOWNHOUSE is the furthest on the right, it will be dropped
df = df.drop('homeType_TOWNHOUSE', axis=1)
```

<h4>Missing handeling</h4>

```python
df = df.dropna(subset=["longitude"])
df = df.dropna(subset=["yearBuilt"])
df = df.dropna(subset=["zestimateLowPercent"])
df = df.dropna(subset=["zestimate"])
df = df.dropna(subset=["propertyTaxRate"])
df = df.dropna(subset=["bathrooms"])
df = df.dropna(subset=["bedrooms"])
df = df.dropna(subset=["zestimateHighPercent"])
df = df.dropna(subset=["livingArea"])

#drop all prices that are 0
df = df[df['price'] != 0]

#use backfill to fill countyFIPS
df['countyFIPS'] = df['countyFIPS'].fillna(method='bfill')

# For timeonZillow, you cannot use ffill or bfill since it is put at random, so imma us a distribution fill
# Fill missing values in 'timeOnZillow' using random sampling from the existing non-missing values
df['timeOnZillow'] = df['timeOnZillow'].apply(
    lambda x: np.random.choice(df['timeOnZillow'].dropna()) if pd.isna(x) else x
)


```
